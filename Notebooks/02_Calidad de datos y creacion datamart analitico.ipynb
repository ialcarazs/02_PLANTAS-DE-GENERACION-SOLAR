## CREACI칍N DEL DATAMART ANALITICO

Vamos a hacer una uni칩n por partes.

Primero los dos datasets de generaci칩n. Que ser치 una apilaci칩n de registros ya que los campos son iguales.

Despu칠s los dos de medidas ambientales. Que ser치 una apilaci칩n de registros ya que los campos son iguales.

Y por 칰ltimo cruzaremos ambos parciales mediante la integraci칩n por campos clave.

### Uni칩n de los datasets de generaci칩n

```python
gener = pd.concat([p1g,p2g],axis = 'index')
gener
```

Vamos a renombrar ya las variables para hacerlas m치s descriptivas y usables.

```python
gener.columns = ['fecha','planta','inverter_id','kw_dc','kw_ac','kw_dia','kw_total']
gener
```

Ahora que tenemos las 2 plantas unidas vamos a hacer lo que se llama un an치lisis de coherencia, dado que seg칰n la documentaci칩n kw_dia y kw_total est치n directamente relacionados con kw_dc y kw_ac.

Vamos a intentar replicar los datos de kw_dia y kw_total.

```python
gener2 = gener.copy()
```

Creamos una variable date para poder agregar por ella.

```python
gener2['date'] = gener2.fecha.dt.date
gener2
```

La suma por planta, date e inverter de kw_dc o de kw_ac deber칤a coincidir con el m치ximo de kw_dia.

```python
gener2 = gener2.groupby(['planta','date','inverter_id']).agg({'kw_dc':sum,
                                                              'kw_ac':sum,
                                                              'kw_dia':max,
                                                              'kw_total':max}).reset_index()
gener2
```

Ordenamos para poder analizar.

```python
gener2 = gener2.sort_values(['planta','inverter_id','date'])
gener2
```

Kw_dia no concuerda para nada ni con kw_dc ni con kw_ac.

Vamos a ver si concuerda con kw_total, para ello calculamos el incremento diario de kw_total que deber칤a coincidir con el m치ximo de kw_dia del d칤a anterior.

```python
gener2['lag1'] = gener2.groupby(['planta','inverter_id']).kw_total.shift(1)
gener2['incremento'] = gener2.kw_total - gener2.lag1
gener2
```

Comprobamos en la planta 1.

```python
gener2[gener2.planta == 'p1'].head(50)
```

Comprobamos en la planta 2.

```python
gener2[gener2.planta == 'p2'].head(50)
```

Conclusiones:
* kw_dia tiene coherencia con kw_total
* pero 칠stas no tienen coherencia con kw_dc ni con kw_ac
* es como si estuvieran en diferentes unidades o hubiera alg칰n c치lculo del que no somos conscientes
* por tanto tendremos 2 bloques a poder usar: o bien kw_dc con kw_ac, o bien kw_dia con kw_total, pero no podemos mezclarlas entre s칤

En este caso el campo clave es compuesto de fecha y planta y manda el dataset de generaci칩n, ya que el de temperatura solo nos aporta variables adicionales.

```python
df = pd.merge(left = gener, right = temper, how = 'left', on = ['fecha','planta'])
df
```

```python
df.isna().sum()
```

Buscamos si los nulos cumplen alg칰n patr칩n.

Se trata del d칤a 3 de Junio a las 14:00, que por alg칰n motivo no tiene datos de temperatura pero solo para 4 inverters de la planta 1.

Vamos a buscar en el dataset de temperatura si existe ese datetime.

```python
temper[temper.fecha.between('2020-06-03 13:30:00', '2020-06-03 14:30:00')]
```

Efectivamente vemos que falta ese tramo en ambas plantas. Pero sin embargo solo hay mediciones en esa hora en la planta 1, y solo en 4 inverters.

Por tanto habr칤a dos soluciones:

* imputar esos datos para esos invertes
* eliminar esos 4 registros

Dado que parece una franja de medici칩n propia solo de 4 inverters de la planta 1 vamos a optar por eliminarlos.

```python
df.dropna(inplace = True)
df
```
