# Creación del datamarkt análitico

```python
gener = pd.concat([p1g, p2g], axis='index')
gener.head()
```

```python
gener.columns = ["fecha", 'planta','inverter_id','kw_ac','kw_dc','kw_dia','kw_total']
gener
```

```python
gener2 = gener.copy()
```

```python
gener2['day'] = gener.fecha.dt.date
gener2
```

```python
gener2 = gener2.groupby(['planta','day','inverter_id']).agg({'kw_dc':sum,
                                                 'kw_ac':sum,
                                                 'kw_dia':max,
                                                 'kw_total': max}).reset_index()
```

```python
gener2 = gener2.sort_values(['planta','inverter_id','day'])
gener2
```

```python
gener2['lag1'] = gener2.groupby(['planta','inverter_id']).kw_total.shift(1)
gener2['incremento'] = gener2.kw_total - gener2.lag1
gener2
```

```python
gener2[gener2.planta == 'p1'].head(50)
```

```python
temper = pd.concat([p1w, p2w], axis= 'index')
temper
```

```python
temper.columns = ['fecha', 'planta', 'sensor_id','t_ambiente','t_modulo','t_irradiacion']
temper
```

```python
df = pd.merge(left= gener, right=temper, how= 'left', on= ['fecha', 'planta'])
df
```
